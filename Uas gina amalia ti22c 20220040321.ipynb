```python
import matplotlib.pyplot as mp
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/ginaamalia/uas-gina-amalia/main/uas_dpp.csv',sep=';')
```

A. Tampilkan semua kolom dengan tambahan description Kolom description akan bernilai "PASS" jika grade >= 70 dan "FAIL" jika grade < 70

```python
#Menambahkan column description berdasarkan nilai dari column Grade
df.loc[(df['Grade'] >= 70),'Description'] = 'PASS' 
df.loc[(df['Grade'] < 70),'Description'] = 'FAIL'

df
```

B.Tampilkan kolom student number,name, dan mean (rata- rata) dari semua nilai perstudent

```python
#opsi 1 jawaban no 2 B
mean_per_student = df.groupby(["Student Numeber", "Name"])["Grade"].mean()

print(mean_per_student)
print(f"rata-rata {df['Grade'].mean()}")
```

```python
#opsi 2 jawaban no 2 B
df['rata-rata'] = df['Grade'].mean()
fl1 = df[['Student Numeber','Name','rata-rata']]
fl1
```

```python
#opsi 3 jawaban no 2 B
df['rata-rata'] = df['Grade'].mean()
df
```

C.Buatlah visualisasi data dari proses nomor 2 dengan hasil seperti berikut

```python
#Visualiasi data dalam bentuk bar

x = df['Name']
y = df['Grade']

mp.bar(x,y)
mp.title('Mean of Final Exam Values')
mp.xlabel('Name')
mp.ylabel('Grade')
mp.show()
```
